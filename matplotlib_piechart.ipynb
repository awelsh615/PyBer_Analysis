```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
categories = ['pants', 'shirts', 'outerwear', 'pajamas', 'underwear', 'shoes']
```

```python
revenue = [25, 15, 5, 20, 10, 25]
```

```python
plt.pie(revenue)
plt.show
```

```python
plt.pie(revenue,labels = categories)
plt.show
```

```python
# Adding autopct adds the percentages on the pie chart
# The '%1.1f%%' puts % labels and one decimal place
plt.pie(revenue, labels = categories, autopct = '%1.1f%%')
plt.show
```

```python
# explode argument - provide list of values between 0 and 1 same length as revenue
explode = [0,.15,0,0,0,0]
plt.pie(revenue, labels = categories, autopct = '%1.1f%%', explode = explode)
plt.show
```

```python
# Matlab method

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Make pie chart bigger so percentages show clearly
plt.subplots(figsize=(8,8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Object oriented approach

fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis)
plt.show()
```

```python
# Skill Drill
# From documentation
# Axes.pie(self, x, explode=None, labels=None, colors=None, 
# autopct=None, pctdistance=0.6, shadow=False, labeldistance=1.1, startangle=0, 
# radius=1, counterclock=True, wedgeprops=None, textprops=None, center=0, 0, 
# frame=False, rotatelabels=False, *, normalize=None, data=None)


# Add a percentage to one decimal place to each wedge of the pie.
# Increase the figure size to 8x8
# Add a shadow
# Explode the two highest percentage months
# Add a start angle so that January is on to
# Reverse the order so that the month order is clockwise
# Add new colors of your own choosing or use the colors from the previous pie chart.

fig, ax = plt.subplots()
ax.pie(y_axis,
       labels=x_axis, 
       shadow = True, 
       radius = 2, 
       startangle = 90, 
       autopct ='%.1f%%',
       explode = [0,0,.3,0,0,0,.3,0,0,0,0,0],
       counterclock = False,
       colors = ['red','darkorange','gold','yellow','lawngreen','mediumseagreen','mediumturquoise','royalblue','blueviolet','mediumpurple','fuchsia','deeppink'],
       )
plt.show()

```
