```python
# Magic command to show plots in same cell as code
%matplotlib inline
```

```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis)
```

```python
# Create the plot.
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
# Skill Drill
# Change the bar colors to magenta.
plt.barh(x_axis, y_axis, color = 'magenta')

# Add a legend for the city of Boston
plt.legend(['Boston'])

# Add a title and axes labels
plt.title('PyBer Fare by Month')
plt.xlabel('Fare ($)')
plt.ylabel('Date')

# Invert the y-axis so that January is at the top.
plt.gca().invert_yaxis()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
# Skill Drill

# Change the bar color to cyan.
# Switch the axis so that the Fare($) data is on the x-axis.
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color='mediumturquoise')

# Add a legend for the city of Chicago.
ax.legend(['Chicago'])

# Add a title and axes labels.
ax.set_title("PyBer Fare by Month")
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')

# Invert the y-axis data.
ax.invert_yaxis()
```
