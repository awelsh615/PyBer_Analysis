```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Create lists for x and y axis data
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
x_axis=['Jan','Feb','March','April','May','June','July','Aug','Sept','Oct','Nov','Dec']
```

```python
# Create the plot using MATLAB method
plt.plot(x_axis,y_axis)
# plt.xlabel('Month')
# plt.ylabel('Balance in $')
# plt.title('Savings Account')
# plt.show()
```

```python
# Create the plot with ax.plt() - object-oriented interface method
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create a quick graph with ax.plt() - object-oriented interface method
ax=plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# plt.xlim('Jan','June')
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
# Change figure size to 20 pixels x 10 pixels
# plt.subplots(figsize=(20,10))
```

```python
# Create a quick graph with ax.plt() - object-oriented interface method
# Change line color to green
# Change the marker to a diamond
# Change line width to 2 points
ax=plt.axes()
ax.plot(x_axis, y_axis,'gd',linewidth=2, linestyle='-')

# Add a legend for Boston
ax.legend(['Boston'])
# Add a title and axes labels
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
# Add a y-axis limit
ax.set_ylim(0,45)
# Add grid lines
ax.grid()
```
